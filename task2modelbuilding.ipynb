```python
import pandas as pd
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import recall_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from imblearn.combine import SMOTEENN
```

```python
df=pd.read_csv("C:\\Users\\Dell\\Desktop\\Machine Learning\\tel_churn.csv")
df.head()
```

```python
df=df.drop('Unnamed: 0',axis=1)
```

```python
x=df.drop('Churn',axis=1)
x
```

```python
y=df['Churn']
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
# Decision Tree Classifier
```

```python
model_dt=DecisionTreeClassifier(criterion = "gini",random_state = 100,max_depth=6, min_samples_leaf=8)
```

```python
model_dt.fit(x_train,y_train)
```

```python
y_pred=model_dt.predict(x_test)
y_pred
```

```python
model_dt.score(x_test,y_test)
```

```python
print(classification_report(y_test, y_pred, labels=[0,1]))
```

```python
from imblearn.combine import SMOTEENN

sm = SMOTEENN()
X_resampled, y_resampled = sm.fit_resample(x, y)
```

```python
xr_train,xr_test,yr_train,yr_test=train_test_split(X_resampled, y_resampled,test_size=0.2)
```

```python
model_dt_smote=DecisionTreeClassifier(criterion = "gini",random_state = 100,max_depth=6, min_samples_leaf=8)
```

```python
model_dt_smote.fit(xr_train,yr_train)
yr_predict = model_dt_smote.predict(xr_test)
model_score_r = model_dt_smote.score(xr_test, yr_test)
print(model_score_r)
print(metrics.classification_report(yr_test, yr_predict))
```

```python
print(metrics.confusion_matrix(yr_test, yr_predict))
```

```python
# Applying PCA
from sklearn.decomposition import PCA
pca = PCA(0.9)
xr_train_pca = pca.fit_transform(xr_train)
xr_test_pca = pca.transform(xr_test)
explained_variance = pca.explained_variance_ratio_
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(
    n_estimators=100,
    criterion='gini',
    random_state=100,
    max_depth=6,
    min_samples_leaf=8
)
```

```python
model.fit(xr_train_pca,yr_train)
```

```python
yr_predict_pca = model.predict(xr_test_pca)
```

```python
model_score_r_pca = model.score(xr_test_pca, yr_test)
```

```python
print(model_score_r_pca)
print(metrics.classification_report(yr_test, yr_predict_pca))
```
